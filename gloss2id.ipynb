```python
!pip install nltk
!python -m nltk.downloader punkt
import nltk
from nltk.tokenize import word_tokenize

# Read sentences from a file (replace 'myfile.txt' with your file path)
with open('Sign sentence.txt', 'r', encoding='utf-8') as fin:
    sentences = fin.read()

# Tokenize the sentences into words
tokens = word_tokenize(sentences)

# Print the tokens
print(tokens)
print(len(tokens))

tokens = list(dict.fromkeys(tokens))
print(tokens)
print(len(tokens))
items  = ['<mask>', '<unk>', '</s>', '<pad>', '<s>']
for item in items:
    tokens.insert(0, item)

print(tokens)
```

```python
alignment_ids = {gloss: i for i, gloss in enumerate(tokens)}

print("Alignment IDs:")
gloss2id={}
for gloss, alignment_id in alignment_ids.items():
    gloss2id[gloss]=alignment_id
    print(f"{gloss}: {alignment_id}")

```

```python
import pickle
print(gloss2id)
with open('/content/gloss2id.pkl', 'wb') as file:
    pickle.dump(gloss2id, file)

```
